### Inspecting GRIB contents

We read a GRIB file containing 6 messages:

```python
import emohawk

fs = emohawk.load_from("file", "test6.grib")
```

#### Using ls()

```python
fs.ls()
```

We can add extra keys:

```python
fs.ls(extra_keys=["paramId"])
```

#### Using describe

We can describe the whole object:

```python
fs.describe()
```

We can alos describe a given parameter (defined by shortName or paramId):

```python
fs.describe("t")
```

```python
fs.describe(131)
```
